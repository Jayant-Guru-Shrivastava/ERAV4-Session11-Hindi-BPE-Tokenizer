```
from bpe.bpe import ByteLevelBPE
import glob

# Read a small sample corpus from data/ (add your own large files for real training)
paths = glob.glob("data/*.txt")
if not paths:
    # create a tiny demo corpus
    import os
    os.makedirs("data", exist_ok=True)
    with open("data/hindi_demo.txt","w",encoding="utf-8") as f:
        f.write("नमस्ते भारत! यह एक छोटा-सा डेमो है। मुंबई एक सुंदर शहर है।\nHello from Mumbai.")

with open("data/hindi_demo.txt","rb") as f:
    corpus = f.read()

bpe = ByteLevelBPE()
bpe.train(corpus, vocab_size=8192, progress=None)
bpe.save("artifacts")

text = "नमस्ते! यह एक हिंदी वाक्य है। Hello from Mumbai!"
ids = bpe.encode(text)
dec = bpe.decode(ids)
print("ids:", ids[:40], "...")
print("decoded:", dec)
cr = len(text.encode('utf-8'))/max(1,len(ids))
print("Compression ratio:", cr)
```
